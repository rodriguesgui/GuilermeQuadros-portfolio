# Vendas Globais
Qual são os anos em que o PS4 mais foi vendido? com uma rápida busca descobrimos que o os anos onde o console mais foi vendido foi em 2015, 2016 e 2017. Outro dado importante é ausência de anos mais recentes, o csv estava com datas até 2020 porém os dados de 2019 e 2020 estavam nulos e não seriam eficientes para a pesquisa

```python
#Retirar os anos com dados nulos
Base_Dados = Base_Dados.loc[ (Base_Dados['Year'] != 2019) & (Base_Dados['Year'] != 2020)]
Base_Dados.head()
```

```python
#Vendas globais de jogos no PS4 ao longo dos anos

plt.figure( figsize = (14,5))
plt.title('Quantidade de vendas globais (MI)', loc='left', fontsize=14)

sns.barplot( data=Base_Dados, x='Year', y='Global', ci=None, color='#69b3a2', estimator=sum)
plt.ylabel('Quantidade vendas (MI)');
```

```python
#Distribuição das vendas globais

plt.figure( figsize = (14,5))

plt.style.use('ggplot')

plt.title('Distribuição das vendas globais', loc='left', fontsize=14)
sns.kdeplot( Base_Dados['Global'], shade=True, bw=1, color='#96a8a8', linewidth=2);
```

# Qual é a distribuição demográfica das vendas?

A seguir vamos apresentar a distribuição das vendas, separando em quatro blocos, América do Norte, Europa, Japão e o Restante do mundo.

```python
Analise = Base_Dados.groupby( by=['Year'] ).sum().reset_index()
```

```python
#Criando a porcentagem de vendas de cada região
America = [ America / Total * 100 for America, Total in zip(Analise['North America'], Analise['Global'])]
Europa = [ Europa / Total * 100 for Europa, Total in zip(Analise['Europe'], Analise['Global'])]
Japao = [ Japan / Total * 100 for Japan, Total in zip(Analise['Japan'], Analise['Global'])]
Mundo = [ Mundo / Total * 100 for Mundo, Total in zip(Analise['Rest of World'], Analise['Global'])]

America, Europa, Japao, Mundo
```

```python
plt.figure(figsize=(16, 5))
Largura_Barra = 0.85
Rotulos = Analise ['Year']
Grupos = [0, 1, 2, 3, 4, 5]
plt.title('Análise distribuição por continente')
#plot da america
plt.bar(Grupos, America, width=Largura_Barra, color='#b5ffb9', edgecolor='white')

#plot da Europa
plt.bar(Grupos, Europa, bottom=America, width=Largura_Barra, color='#f9bc86', edgecolor='white')

#plot da Japan
plt.bar(Grupos, Japao, bottom=[A + B for A, B in zip(America, Europa)], width=Largura_Barra, color='#f87778', edgecolor='white')

#plot do Resto do mundo
plt.bar(Grupos, Mundo, bottom=[A + B + C for A, B, C in zip(America, Europa, Japao)], width=Largura_Barra, color='#d3acfe', edgecolor='white')
 #Labels
plt.xticks( Grupos, Rotulos)
plt.xlabel('Grupo')
plt.ylabel('Distribuição %')

#legenda

plt.legend(['America do Norte', 'Europa', 'Japão', 'Mundo'], loc='upper left', bbox_to_anchor=(0.26, -0.1), ncol=4);
```

# Resultado:

POodemos notar que a Europa é a região onde mais foi vendido o console durante os anos de 2013 a 2017 ficando de fora apenas em 2018 quando a Região da America do Norte ficou em primeiro

# Quais foram as empresas que mais venderam? Qual foi o gênero mais comprado? E qual foi o jogo mais vendido?"

```python
#Tranformando dados em números para facilitar o acesso aos dados
from sklearn.preprocessing import LabelEncoder
Funcao_label = LabelEncoder()

Base_Dados['Produtor'] = Funcao_label.fit_transform(Base_Dados['Publisher'])
Base_Dados['Genero'] = Funcao_label.fit_transform(Base_Dados['Genre'])
Base_Dados['Jogo'] = Funcao_label.fit_transform(Base_Dados['Game'])

Base_Dados.head()
```

```python
#Paleta de cores para usar nos gráficos
Paleta_Cores = sns.color_palette('Spectral', 8)

Paleta_Cores
```

# A Empresa que mais vendeu?

Ao analisarmos o gráfico, podemos notar que duas empresa tem maior destaque no gráfico, estas empresas são Activision e a Rockstar Games

```python
#A Empresa que mais vendeu?
plt.figure(figsize=(20, 5))
plt.title('Analise dos produtores de Jogos (MI)')
sns.scatterplot(data = Base_Dados, x='Produtor', y='Global', color=Paleta_Cores[0]);
```

# O gênero mais jogado

Os gêneros mais jogados de acordo com o gráfico são os números 0, 1, 12 e 14
respectivamente correspondem aos gêneros de ação, ação e aventura, tiros e esportes.

```python
#O gênero mais jogado
plt.figure(figsize=(20, 5))
plt.title('Analise dos produtores de Jogos (MI)')
sns.scatterplot(data = Base_Dados, x='Genero', y='Global', color=Paleta_Cores[1]);
```

# O jogo mais jogado

Nesse gráfico notamos um jogo em destaque, o jogo mais jogado foi Grand Theft Auto V

```python
# o jogo mais jogado
plt.figure(figsize=(20, 5))
plt.title('Analise dos produtores de Jogos (MI)')
sns.scatterplot(data = Base_Dados, x='Jogo', y='Global', color=Paleta_Cores[2]);
```

# Preferências Regionais por Gênero de Jogo

Com estes dados conseguimos ter uma noção dos gêneros mais jogados por região, nota-se que os gêneros mais jogados nas regiões da Europa e America do Norte são os mesmos, estes também são os mais jogados em todo o resto do mundo

```python


# Vamos considerar apenas as colunas relevantes para o gráfico
cols = ['North America', 'Europe', 'Japan', 'Rest of World', 'Genre']

# Filtrando o DataFrame para incluir apenas as colunas relevantes
df_filtered = Base_Dados[cols]

# Agrupando os dados pelo gênero e somando as preferências regionais para cada gênero
grouped_df = df_filtered.groupby('Genre').sum()

# Criando o gráfico de barras empilhadas
grouped_df.plot(kind='bar', stacked=True, figsize=(12, 6))

# Adicionando rótulos e título ao gráfico
plt.xlabel('Gênero')
plt.ylabel('Preferências Regionais')
plt.title('Preferências Regionais por Gênero de Jogo')

# Exibindo o gráfico
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
